### Lets do univarite analysis

```python
def univariate(df ,column):
    plt.figure()
    ax=sns.countplot(data=df,x=column,palette='Set2')
    total=df.shape[0]
    
    for p in ax.patches:
        height=p.get_height()
        ax.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:1.2f} %'.format((height/total)*100),
            ha="center") 
    plt.title("count and % of {} feature".format(column))
```

#### starting with categorical feature

```python
def bivariate(df ,column,target):
    plt.figure()
    ax=sns.countplot(data=df,x=target,hue=column,dodge=True,palette="husl")
    total=df.shape[0]
    
    for p in ax.patches:
        height=p.get_height()
        ax.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:1.2f} %'.format((height/total)*100),
            ha="center") 
    plt.title("count and % of {} feature".format(column))
```

```python
def cont_bivartiate(df,column,target,mval):
    #
    #plt.scatter(x=column,y=target,data=df)
    plt.figure(figsize=(12,8))
    sns.stripplot(x=column,y=target,data=df,jitter=True,palette='husl')
    plt.xlim(0,mval)
    plt.title("Loan_status and {} plot".format(column))
```

```python
def calculate_missing_percent(df):
    missing=pd.DataFrame()
    percentage=[]
    count=[]
    total_record=len(df)
    for feature in df.columns:
        cnt=df[feature].isna().sum()
        cal=cnt/total_record
        percentage.append(cal*100)
        count.append(cnt)
    missing['feature']=df.columns
    missing['count of missing values']=count
    missing['missing %']=percentage
    return missing
        
    
```

```python
calculate_missing_percent(train)
```
